**Partie 1 : Modélisation des faits et règles**

```python
#Definition des faits 
fait1 = "fievre"
fait2 = "toux"
fait3 = "douleurs_thoraciques"
fait4 = "essoufflement"
```

```python
# Définition des règles sous forme de dictionnaires

R1 = {
    "conditions": ["fievre", "toux"],
    "conclusion": "grippe"
}

R2 = {
    "conditions": ["grippe", "douleurs_thoraciques"],
    "conclusion": "infection_respiratoire"
}

R3 = {
    "conditions": ["infection_respiratoire", "essoufflement"],
    "conclusion": "hospitalisation_conseillee"
}

# Base de règles
base_de_regles = [R1, R2, R3]

base_de_regles
```

**Partie 2 : Moteur d’inférence en chaînage avant**

```python
def chainage_avant(faits_initiaux, regles):

    faits = set(faits_initiaux)   
    nouveaux_faits = True         
    while nouveaux_faits:
        nouveaux_faits = False

        for regle in regles:
            conditions = regle["conditions"]
            conclusion = regle["conclusion"]

            # Vérifier si toutes les conditions de la règle sont satisfaites
            if all(cond in faits for cond in conditions):
                
                # Si la conclusion n'est pas déjà un fait, on l'ajoute
                if conclusion not in faits:
                    faits.add(conclusion)
                    nouveaux_faits = True  

    return list(faits)
```

Cas 1 : 

```python
faits_initiaux_1 = ["fievre", "toux"]

resultat_1 = chainage_avant(faits_initiaux_1, base_de_regles)
resultat_1
```

Cas 2 : 

```python
faits_initiaux_2 = ["fievre", "toux", "douleurs_thoraciques", "essoufflement"]

resultat_2 = chainage_avant(faits_initiaux_2, base_de_regles)
resultat_2
```

Cas 3 : 

```python
faits_initiaux_3 = ["fievre"]

resultat_3 = chainage_avant(faits_initiaux_3, base_de_regles)
resultat_3
```

**Partie 3 : Interaction avec le système**

```python
def peut_on_deduire(fait_cible, faits_initiaux, regles):
    # On applique le chaînage avant pour obtenir tous les faits déduits
    faits_finaux = chainage_avant(faits_initiaux, regles)

    # Vérification de la présence du fait cible
    if fait_cible in faits_finaux:
        print(f"Oui, {fait_cible} est déduit.")
        return True
    else:
        print(f"Non, le système ne peut pas conclure {fait_cible}.")
        return False
```
